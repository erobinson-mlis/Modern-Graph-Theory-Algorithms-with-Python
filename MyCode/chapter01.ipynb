```python
#import igraph
import igraph as ig
from igraph import Graph
```

```python
# create hackathon network
g_colleagues = ig.Graph(
    edges=[(0,1), (0,2), (1,2)], n=3, directed=False)
```

```python
# plot the hackathon network
ig.plot(g_colleagues, bbox=(200,200), vertex_size=40, 
        vertex_label=["M","A","G"])
```

```python
# install networkx
import networkx as nx
```

```python
# create hackathon network
G = nx.Graph()
G.add_nodes_from([1,3])
G.add_edges_from([(1,2), (1,3), (2,3)])

# plot the hackathon network
import matplotlib.pyplot as plt
G.nodes[1]['initial'] = 'M'
G.nodes[2]['initial'] = 'A'
G.nodes[3]['initial'] = 'G'
labels = nx.get_node_attributes(G, 'initial')
nx.draw(G, labels=labels, font_weight='bold')
```

```python
# Create an Erdos-Renyi network
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1,2)

ax1.set_title("Erdos-Renyi, p=0.70")
g1 = ig.Graph.Erdos_Renyi(n=20, p=0.7, directed=False, loops=False)
ig.plot(g1, target=ax1, bbox=(500,500),vertex_size=20)

ax2.set_title("Erdos-Renyi, p=0.10")
g2 = ig.Graph.Erdos_Renyi(n=20, p=0.1, directed=False, loops=False)
ig.plot(g2, target=ax2, bbox=(500,500), vertex_size=20)
```

```python
fig, axs = plt.subplots(1,2, figsize=(10,5))

axs[0].set_title("Erdos-Renyi, p=0.70")
g1 = nx.erdos_renyi_graph(n=20, p=0.70)
nx.draw(g1, ax=axs[0])

axs[1].set_title("Erdos-Renyi, p=0.10")
g2 = nx.erdos_renyi_graph(n=20, p=0.10)
nx.draw(g2, ax=axs[1], with_labels=True)

plt.show()
```
